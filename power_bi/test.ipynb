```python
#superhost and review score rating
df['host_is_superhost'].value_counts()
df['review_scores_rating'].mean()
```

```python
#compare the reviezs average for superhost and non superhost
df.groupby('host_is_superhost')['review_scores_rating'].mean()

#graphique pour montre la difference
import matplotlib.pyplot as plt
df.groupby('host_is_superhost')['review_scores_rating'].mean().plot(kind='bar')
plt.legend()
#rajouter valeurs dans les colonnes
for i, v in enumerate(df.groupby('host_is_superhost')['review_scores_rating'].mean()):
    plt.text(i, v+0.5, str(round(v, 2)), ha='center')

    
plt.show()
```

```python
#comparere la quantité de review pour superhost et non superhost
df.groupby('host_is_superhost')['number_of_reviews'].mean()

df.groupby('host_is_superhost')['number_of_reviews'].mean().plot(kind='bar')
plt.legend()
for i, v in enumerate(df.groupby('host_is_superhost')['number_of_reviews'].mean()):
    plt.text(i, v+0.5, str(round(v, 2)), ha='center')
plt.show()
```

```python
#avoir plus de reviews permet de louer plus des chambres
df.groupby('host_is_superhost')['number_of_reviews'].mean().plot(kind='bar')
plt.legend()
for i, v in enumerate(df.groupby('host_is_superhost')['number_of_reviews'].mean()):
    plt.text(i, v+0.5, str(round(v, 2)), ha='center')
plt.show()
```
